```python
import os
import PIL.ImageOps    
from PIL import Image
from tqdm.notebook import tqdm

save_path = "SeekShot3Invert/"
!mkdir {save_path}


version = "SeekShot3/"
folder_list = os.listdir(version)
for folder in folder_list:
    file_list = os.listdir(version + folder)
    folder += "/"
    !mkdir {save_path + folder}
    for file in tqdm(file_list):
        im = Image.open(version + folder + file)
        im = PIL.ImageOps.invert(im)
        im.save(save_path + folder + file)

```
