# unzip

```python
import os

foldername = "SeekShot3/"

file_list = os.listdir(foldername)

for file in file_list:
    dirname = file.split('.')[0]
    print(dirname)
    !unzip {foldername}{file} -d {foldername}{dirname}
    
```

# jpg to png

```python
import os
from PIL import Image

file_list = os.listdir(foldername)

for i, file in enumerate(file_list):
    
    jpg_dir = foldername + file + "/"
    jpg_list = os.listdir(jpg_dir)
    for jpg in jpg_list:
        png = jpg.split('.')[0] + ".png"
        im = Image.open(jpg_dir + jpg)
        im.save(jpg_dir + png)
    !rm {jpg_dir}*.jpg
        
    jpg_list = os.listdir(foldername + file)
    print(jpg_list[:10])
```
